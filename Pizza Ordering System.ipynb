```python
# def to choose from options
def option_choose(options, name):
    index = 0
    indexValidList = []
    print('Select a ' + name + ':')
    for optionName in options:
        index = index + 1
        indexValidList.extend([optionName])
        print(str(index) + ') ' + optionName)
    inputValid = False
    while not inputValid:
        inputRaw = input(name + ': ')
        inputNo = int(inputRaw) - 1
        if inputNo > -1 and inputNo < len(indexValidList):
            selected = indexValidList[inputNo]
            print('Selected ' +  name + ': ' + selected)
            inputValid = True
            break
        else:
            print('Please select a valid ' + name + ' number')

    return selected
```

### SORTING

```python
def quicksort(my_list, index):
    if len(my_list) == 0:
        return []
    else:
        pivot = my_list[0]
        lesser = quicksort([x for x in my_list[1:] if x[index] < pivot[index]], index)
        greater = quicksort([x for x in my_list[1:] if x[index] >= pivot[index]], index)
        return lesser + [pivot] + greater
```

### Complexity

Algorithmic Complexity
Sorting used is quick sort which is a divide and conquer algorithm. It picks an element as pivot and partitions the given array around the picked pivot.The key process in quickSort is partition(). Target of partitions is, given an array and an element x of array as pivot, put x at its correct position in sorted array and put all smaller elements (smaller than x) before x, and put all greater elements (greater than x) after x. All this should be done in linear time. 

Time Complexity:With a good pivot, the Quick Sort function would partition the array into halves which grows logarithmically with n. Therefore the average time complexity of the Quick Sort algorithm is O(nlog(n)).
Space Complexity :  This quick sort has a space complexity of O(log n).

Searching Algorithm : In this project we used a linear searching algorithm whose time complexity is O(n) if suppose the element we are searching for is in the first position of the list then it will be O(1) which is its best case scenario. Space complexity here we don't need any extra space to store anything. We just compare the given value with the elements in a list one by one it is O(1).
